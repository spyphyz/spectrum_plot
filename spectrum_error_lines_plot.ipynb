```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
path = 'Dropbox/OpenUniversity/S382_Astrophysics/SDSS_Project/sdss_abell/spectrum_z03_z18/master_z03_z18/auto_z03_z18'
file = 'spectrum_z03_z18_auto.csv' #this would be the SpecCombine csv output file
df = pd.read_csv(os.path.join(path, file))
```

```python
df['SD']=df['error']*df['data_points']**0.5
df['SNR']=df['flux']/df['SD']
s = 1 #s=SNR_scaling_factor
df['SNR'] = df['SNR']*s
df['+SEM'] = df['flux']+df['error']
df['-SEM'] = df['flux']-df['error']
df
```

```python
line_1 = 1549; # y1 =  # C IV emission line
line_2 = 2798; # y1 =  # Mg II emission line
line_3 = 4861; # y1 =  # H-beta I emission line
line_4 = 4340; # y1 =  # H-gamma emission line
line_5 = 1908; # y1 =  # CIII] emission line - semi-forbidden line
line_6 = 1396; # y1 =  # Si IV emission line
line_7 = 5006; # y1 =  # [O III] emission line - semi-forbidden line
line_8 = 3203; # y1 =  # He II emission line
line_9 = 2326; # y1 =  # C II] emission line - semi-forbidden line
line_10 = 2418; # y1 =  # [Ne IV] emission line - semi-forbidden line
line_11 = 5875; # y1 =  # He I 
line_12 = 4101; # y1 =  # H-delta emission line
line_13 = 3726; # y1 =  # [O II] forbidden emission
line_14 = 3868; # y1 =  # [Ne III] emission line - forbidden

wave = df['wavelength']
err_1 = df['flux']+df['SD'] # +SD
err_2 = df['flux']-df['SD'] # -SD
# err_3 = df['flux']+*df['error'] # +SEM
# err_4 = df['flux']-*df['error'] # -SEM

ax1 = df.plot(x='wavelength', y='flux', lw=0.3, color='red', label='mean') 

ax2 = df.plot(x='wavelength', y='+SEM', lw=0.000001
             , color='yellow', label='Â±SD', ax=ax1)

# ax3 = df.plot(x='wavelength', y ='+SEM', lw=0.05
 #            , color='grey', label='Â±SD', ax=ax1)

#ax2 = df.plot(x='wavelength', y='SNR', lw=0.2
#             , color='red', ax=ax1)

plt.legend()
plt.title('Composite Spectrum Of Broadline QSO, z = 0.3 - 1.8', fontsize=13, color='black')
# plt.grid()

plt.fill_between(wave, err_1, err_2, color='#CCCCCC') # Â±SD
# plt.fill_between(wave, err_3, err_4, color='grey') # Â±SEM

# ax1.set_facecolor('black') # creates a black background to the plot

ax1.text(0.82, 0.75, '(n = 297)',
        verticalalignment='bottom', horizontalalignment='left',
        transform=ax1.transAxes,
        color='blue', fontsize=10)

# emission & absorption line labelling
plt.vlines(x=line_1, ymin=0, ymax=480, label='line_1', linestyles='dashed', color='purple', lw=0.2)
plt.text(line_1, 500, 'C IV', color='purple', fontsize=5, rotation=45)

plt.vlines(x=line_2, ymin=0, ymax=480, label='line_1', linestyles='dashed', color='purple', lw=0.2)
plt.text(line_2, 500,'Mg II', color='purple', fontsize=5, rotation=45)

plt.vlines(x=line_3, ymin=0, ymax=480, label='line_1', linestyles='dashed', color='purple', lw=0.2)
plt.text(line_3, 490, r'H$\beta$', color='purple', fontsize=5, rotation=45)

plt.vlines(x=line_4, ymin=0, ymax=480, label='line_1', linestyles='dashed', color='purple', lw=0.2)
plt.text(line_4, 495, r'H$\gamma$', color='purple', fontsize=5, rotation=45)

plt.vlines(x=line_5, ymin=0, ymax=480, label='line_1', linestyles='dashed', color='purple', lw=0.2)
plt.text(line_5, 500,'CIII]', color='purple', fontsize=5, rotation=45)

plt.vlines(x=line_6, ymin=0, ymax=500, label='line_1', linestyles='dashed', color='purple', lw=0.2)
plt.text(line_6, 518,'Si IV', color='purple', fontsize=5, rotation=45)

plt.vlines(x=line_7, ymin=0, ymax=460, label='line_1', linestyles='dashed', color='purple', lw=0.2)
plt.text(line_7, 480,'[O III]', color='purple', fontsize=5, rotation=45)

plt.vlines(x=line_8, ymin=0, ymax=480, label='line_1', linestyles='dashed', color='purple', lw=0.2)
plt.text(line_8, 500,'He II', color='purple', fontsize=5, rotation=45)

plt.vlines(x=line_9, ymin=0, ymax=480, label='line_1', linestyles='dashed', color='purple', lw=0.2)
plt.text(line_9, 500, 'C II]', color='purple', fontsize=5, rotation=45)

plt.vlines(x=line_10, ymin=0, ymax=460, label='line_1', linestyles='dashed', color='purple', lw=0.2)
plt.text(line_10, 485,'[Ne IV]', color='purple', fontsize=5, rotation=45)

plt.vlines(x=line_11, ymin=0, ymax=340, label='line_1', linestyles='dashed', color='purple', lw=0.2)
plt.text(line_11, 355,'He I', color='purple', fontsize=5, rotation=45)

plt.vlines(x=line_12, ymin=0, ymax=460, label='line_1', linestyles='dashed', color='purple', lw=0.2)
plt.text(line_12, 475, r'H$\delta$', color='purple', fontsize=5, rotation=45)

plt.vlines(x=line_13, ymin=0, ymax=460, label='line_1', linestyles='dashed', color='purple', lw=0.2)
plt.text(line_13, 480,'[O II]', color='purple', fontsize=5, rotation=45)

plt.vlines(x=line_14, ymin=0, ymax=340, label='line_1', linestyles='dashed', color='purple', lw=0.2)
plt.text(line_14, 360, '[Ne III]', color='purple', fontsize=4, rotation=45)

plt.xlim(1250, 6550)
plt.ylim(0,530)

plt.xlabel(r'Emitted Wavelength / $\AA$',  fontsize=13, color='grey')
plt.ylabel('Spectral Flux Density / Arbitrary Units', fontsize=10, color='grey')

plt.savefig(os.path.join(path, 'spectrum_z03_z18_auto.png'), dpi = 1080)
#plt.savefig(os.path.join(path, 'spectrum_z03_z18_auto.pdf'))
```
